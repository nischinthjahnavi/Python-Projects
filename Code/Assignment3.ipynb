```python
df=pd.read_sql_query("SELECT * FROM vgsales", conn)
df
```

```python
query1 = pd.read_sql('''SELECT 
    CASE
        WHEN year <= 2005 THEN 'pre-2005'
        ELSE 'post-2005'
    END Sales_Year,
    AVG(Global_Sales) AS Average_Global_Sales
FROM
    vgsales
GROUP BY Sales_year ORDER BY Average_Global_Sales DESC''', conn)
```
